# Cluster Analysis

The k-means clustering method is an unsupervised machine learning technique used to identify clusters of data objects in a dataset.

 The first step is to randomly select k centroids, where k is equal to the number of clusters you choose. Centroids are data points representing the center of a cluster.

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Example 01

```python
X = np.array([[1, 2], [1, 4], [1, 0],
              [10, 2], [10, 4], [10, 0]])
kmeans2 = KMeans(n_clusters=2)
```

```python
#Fit the Model;
y_kmeans2 = kmeans2.fit_predict(X)
print(y_kmeans2)
```

```python
kmeans2.predict([[0, 0], [12, 3]])
```

```python
#Centroids of clusters:
kmeans2.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c = y_kmeans2, cmap='rainbow')
```

# Example 02

```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#import the dataset
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data,columns=[iris.feature_names])
df.head()
```

```python
df.isnull().sum()
```

Now we select all four features (sepal length, sepal width, petal length, and petal width) of the dataset in a variable called x so that we can train our model with these features. For this, we use the iloc function on df, and the column index (0,1,2,3) for the above four columns are used, as shown below:

```python
x = df.iloc[:, [0,1,2,3]].values
print(x)
type(x)
```

# The elbow technique for finding the number of clusters¶


To start, let’s arbitrarily assign the value of k as 5. We will implement k-means clustering using k=5. For this we will instantiate the KMeans class and assign it to the variable kmeans5:

```python
# Defining the number of clusters an
kmeans5 = KMeans(n_clusters=5)
```

```python
#Output of Kmeans clustering with value 5
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)
```

```python
#Print the centers of 5 clusters
kmeans5.cluster_centers_
```

```python
# Let's see how many data points are in these 3 clusters.

frame = pd.DataFrame(x)
frame['cluster'] = y_kmeans5
frame['cluster'].value_counts()
```

```python
#Visualize the data

import matplotlib.pyplot as plt
x=iris.data
plt.scatter(x[:,0], x[:,1], c = y_kmeans5, cmap='rainbow')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
```

```python
plt.scatter(x[:,2], x[:,3], c = y_kmeans5, cmap='rainbow')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
```
