```python
import numpy as np
import pandas as pd
data2 = pd.read_csv("diabetes.csv")
data2.head(20)
```

```python
y = data2["Outcome"]
y
```

```python
x = data2.iloc[:,:-1]
x
```

```python
# data set into two parts
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 1)
```

```python
# import the regressor, classifier, confusion_matrix and accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
 
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
 
# fit the regressor with x and y data
regressor.fit(x, y) 
```

```python
model_rfc = RandomForestClassifier()
model_rfc.fit(x_train,y_train)
```

```python
train_acc = model_rfc.score(x_train,y_train)
print("The accuracy for train set is {}".format(train_acc*100))
```

```python
y_pred = model_rfc.predict(x_test)
```

```python
test_acc = accuracy_score(y_test,y_pred)
print("The accuracy for test set is {}".format(test_acc*100))
```

```python
confusion_matrix = confusion_matrix(y_test,y_pred)
print(confusion_matrix) 
```
